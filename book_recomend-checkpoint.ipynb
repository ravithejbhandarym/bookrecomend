```python
#Collaborative Filtering Based Recommender System
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
knowledgable_users= x[x].index
```

```python
knowledgable_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(knowledgable_users)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.head()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
   
    # Step 1: Enumerate the similarity scores
    enumerated_scores = enumerate(similarity_scores[index])
    
# Step 2: Convert to list
    enumerated_list = list(enumerated_scores)
   
# Step 3: Sort by similarity score in descending order
    sorted_items = sorted(enumerated_list, key=lambda x: x[1], reverse=True)
    
# Step 4: Select top similar items (excluding the first one, often the input itself)
    similar_items = sorted_items[1:5]
    
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
g= filtered_rating.groupby('Book-Author').count()['Book-Rating']>=50
famous_author= g[g].index
```

```python
famous_author
```

```python
final_author_ratings = filtered_rating[filtered_rating['Book-Author'].isin(famous_author)]
```

```python
final_author_ratings.head()
```

```python
author_pt = final_author_ratings.pivot_table(index='Book-Author',columns='User-ID',values='Book-Rating')
```

```python
author_pt.head()
```

```python
author_pt.fillna(0,inplace=True)
```

```python
author_pt
```

```python
author_similarity_scores = cosine_similarity(author_pt)
```

```python
author_similarity_scores
```

```python
def recommend_based_author(book_author):
    # Find the index of the given author
    index = np.where(author_pt.index == book_author)[0][0]
    
    # Step 1: Enumerate the similarity scores
    enumerated_scores = enumerate(author_similarity_scores[index])
    
    # Step 2: Convert to list
    enumerated_list = list(enumerated_scores)
    
    # Step 3: Sort by similarity score in descending order
    sorted_items = sorted(enumerated_list, key=lambda x: x[1], reverse=True)
    
    # Step 4: Select top similar items 
    similar_items = sorted_items[:5]
    
    data = []
    for i in similar_items:
        similar_author = author_pt.index[i[0]]
        temp_df = books[books['Book-Author'] == similar_author]
        
        for _, row in temp_df.drop_duplicates('Book-Title').iterrows():
            item = []
            item.append(row['Book-Title'])
            item.append(row['Book-Author'])
            item.append(row['Image-URL-M'])
            data.append(item)
    
    return data
```

```python
recommend_based_author('Anne Rice')
```
